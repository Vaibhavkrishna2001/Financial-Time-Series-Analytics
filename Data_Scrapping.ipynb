## Dataset Cleaning

### Identify the Bad Rows

```python
import pandas as pd

df = pd.read_csv("all_bank_stock_data.csv")
```

```python
df[df['Date'].isna()]
```

## Removing bad rows

```python
df = df.dropna(
    subset=['Date']
)
```

### Verifing

```python
df[df['Date'].isna()]
```

```python
print(df.shape)
```

## Converting date to datetime

```python
df['Date'] = pd.to_datetime(df['Date'])
```

### verify

```python
print(df.dtypes)
```

### Convert Numeric Columns

```python
numeric_cols = [
    'Open',
    'High',
    'Low',
    'Close',
    'Volume'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col])
```

```python
print(df.dtypes)
```

### Check Missing Values

```python
print(df.isnull().sum())
```

### Checking Duplicates

```python
duplicates = df.duplicated().sum()

print("Duplicates:", duplicates)
```

### Logical Consistency Checks

```python
print((df['High'] >= df['Open']).all())

print((df['High'] >= df['Close']).all())

print((df['Low'] <= df['Open']).all())

print((df['Low'] <= df['Close']).all())
```

### checking What caused the False

```python
print("High < Open :", (df['High'] < df['Open']).sum())
print("High < Close:", (df['High'] < df['Close']).sum())
print("Low > Open  :", (df['Low'] > df['Open']).sum())
print("Low > Close :", (df['Low'] > df['Close']).sum())
```

```python
df[df['High'] < df['Open']].head()
```

```python
df[df['Low'] > df['Close']].head()
```

Logical consistency checks were performed to verify that the highest trading price was greater than or equal to the opening and closing prices, and that the lowest trading price was less than or equal to the opening and closing prices. A small number of observations (fewer than 0.05% of the dataset) violated these conditions. Inspection revealed that the differences were very small and are likely attributable to data provider rounding or adjustment procedures. Given the negligible proportion of affected records, these observations were retained in the dataset and were not considered to materially impact subsequent analysis.

## Exploratory Data Analysis

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))

for bank in df['Bank'].unique():

    temp = df[df['Bank'] == bank]

    plt.plot(
        temp['Date'],
        temp['Close'],
        label=bank
    )

plt.title(
    'Closing Stock Price Comparison'
)

plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.legend()

plt.show()
```

## Trading Volume

```python
plt.figure(figsize=(14,6))

for bank in df['Bank'].unique():

    temp = df[df['Bank'] == bank]

    plt.plot(
        temp['Date'],
        temp['Volume'],
        label=bank
    )

plt.title(
    'Trading Volume Comparison'
)

plt.xlabel('Date')
plt.ylabel('Volume')

plt.legend()

plt.show()
```

## Distribution of Closing Prices

```python
plt.figure(figsize=(12,6))

for bank in df['Bank'].unique():

    temp = df[df['Bank'] == bank]

    plt.hist(
        temp['Close'],
        bins=30,
        alpha=0.5,
        label=bank
    )

plt.legend()

plt.title(
    'Distribution of Closing Prices'
)

plt.show()
```

## Volatility Analysis

### Daily Returns

```python
df = df.sort_values(
    ['Bank','Date']
)

df['Daily_Return'] = (
    df.groupby('Bank')['Close']
      .pct_change()
)
```

### Volatility Calculation

```python
volatility = df.groupby(
    'Bank'
)['Daily_Return'].std()

print(volatility)
```

### Volatility Bar Chart

```python
volatility.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
    'Volatility Comparison'
)

plt.ylabel(
    'Standard Deviation of Daily Returns'
)

plt.show()
```
